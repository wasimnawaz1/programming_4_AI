### Combining Data

The `combine_first()` function is used when we have **two datasets with overlapping indices** and want to keep values from one dataset for overlapping indices, and keep values from the other dataset for non-overlapping indices.

This cannot be obtained with merging or concatenation.

````python
myseries1 = pd.Series([50, 40, 30, 20, 10], index=[1,2,3,4,5])
myseries1
````

````python
myseries2 = pd.Series([100, 200, 300, 400], index=[3,4,5,6])
myseries2
````

````python
# Keep values from myseries1 for overlapping indices
print(myseries1, '\n\n\n', myseries2, '\n\n\n',)

myseries1.combine_first(myseries2)
````

````python
# Keep values from myseries2 for overlapping indices
print(myseries1, '\n\n\n', myseries2, '\n\n\n',)

myseries2.combine_first(myseries1)
````

### Handling Outliers

**Outliers** are values outside the expected range of a feature. Common causes include:
- Human errors during data entry
- Measurement (instrument) errors
- Experimental errors during data extraction or manipulation
- Intentional errors to test accuracy of outlier detection methods

A good approach to detect numeric outliers is to use **InterQuartile Range (IQR)** — the difference between the third (Q3) and first (Q1) quartiles.

Using a typical interquartile multiplier $k = 1.5, 2$ or a bigger value.
- **Lower limit** = $Q1 − (\text{IQR} × 1.5)$
- **Upper limit** = $Q3 + (\text{IQR} × 1.5)$

````python
student_frame = pd.DataFrame({'Student Name': ['A','B','C','D','E','F','G'],
    'Sex': ['M','F','M','F','F','M','M'],
    'Age': [10, 14, 60, 15, 16, 15, 11],
    'School': ['Primary','High','High','High','High','High','Primary']
    })
student_frame
````

````python
# Describe important statistics
student_frame.describe()
````

````python
# Calculate lower and upper limits using IQR
Q1 = student_frame['Age'].quantile(0.25)  # 25% (quartile 1)
Q3 = student_frame['Age'].quantile(0.75)  # 75% (quartile 3)
IQR = Q3 - Q1  # InterQuartile Range (IQR)
IQR_mult = IQR * 1.5
lower = Q1 - IQR_mult
upper = Q3 + IQR_mult
print("The lower limit is = ", lower)
print("The upper limit is = ", upper)
````

````python
# Filter the DataFrame to remove outliers
student_frame = student_frame[student_frame['Age'] > int(lower)]
student_frame = student_frame[student_frame['Age'] < int(upper)]
student_frame
````

### Data Mapping

The Pandas library provides useful **data mapping** functions. Mapping is the creation of a list of matches between two values using dictionary objects.

The following functions accept a dictionary object to perform mapping:
- `replace()` — replaces values
- `map()` — creates a new column
- `rename()` — replaces the index values

````python
data = {'color': ['blue','green','yellow','red','white'],
    'object': ['ball','pen','pencil','paper','mug'],
    'price': [1.2,1.0,0.6,0.9,1.7]}
myframe = pd.DataFrame(data)
myframe
````

````python
# Create a dictionary for mapping and use replace()
mymap = {'blue': 'dark blue', 'green': 'light green'}
myframe.replace(mymap)
````

````python
# Use replace() to replace NaN values
myseries = pd.Series([1, 2, np.nan, 4, 5, np.nan])
print(myseries, '\n\n\n')
myseries.replace(np.nan, 0)
````

````python
# Use map() to add a new column to the DataFrame
mymap2 = {'ball':'round', 'pencil':'long', 'pen':'long', 'mug':'cylindrical', 'paper':'rectangular'}
myframe['shape'] = myframe['object'].map(mymap2)
myframe
````

````python
# Use rename() to rename indices
reindex = {0: 'first', 1: 'second', 2: 'third', 3: 'fourth', 4: 'fifth'}
myframe = myframe.rename(reindex)
myframe
````

### Discretization and Binning

**Discretization** transforms continuous data into **discrete categories** (bins) to facilitate analysis. For example, dividing a range of values into intervals.

- `pd.cut()` — divides data into **equal-width** bins
- `pd.qcut()` — divides data into **quantile-based** bins (equal number of occurrences per bin)

````python
readings = [34, 39, 82, 75, 16, 17, 15, 74, 37, 68, 22, 92, 99, 54, 39, 96, 17, 36, 91, 86]

bins = [0, 25, 50, 75, 100]
mycategory = pd.cut(readings, bins)
mycategory
#type(mycategory)   # check the type of mycategory
````

````python
readings.sort()
print(readings)
````

````python
# Count the number of elements in each bin
pd.value_counts(mycategory)
````

````python
# Give meaningful names to the bins
bin_names = ['Poor', 'Below Average', 'Average', 'Good']
pd.cut(readings, bins, labels=bin_names)
````

````python
# qcut(): quantile-based binning (equal number of elements per bin) qcut uses the formula cut = (n-1)*p ---> n=20, p=0.25, 0.5, 0.75, 1
cut_values = pd.qcut(readings, 4)

print(cut_values)
````

````python
# Check the number of elements in each qcut bin
pd.value_counts(pd.qcut(readings, 4))
````

### Aggregating Data

**Aggregation** is the process of grouping data together using statistical functions such as mean, median, count, or sum to combine several rows. The combined data is:
- Easier to analyze
- Protects individual privacy
- Can be matched with other data sources

````python
data = {'color': ['blue','white','red','red','white'],
    'object': ['ball','pen','pencil','paper','mug'],
    'price': [1.2,1.0,0.6,0.9,1.7]}
myframe = pd.DataFrame(data)
myframe
````

````python
# Group data based on the column 'color'
mygroup = myframe['price'].groupby(myframe['color'])
print(mygroup.groups)
````

````python
mygroup.sum()
````

````python
# Q8

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
print("Q8:\n", df.describe())

# Q9
print("Q9:\n", df[df['A']>1])

# Q10
df['C'] = df['A'].apply(lambda x: x*2 if x>1 else x)
print("Q10:\n", df)

# Q11
df2 = pd.DataFrame({'grp':['A','A','B'],'val':[10,20,30]})
print("Q11:\n", df2.groupby('grp')['val'].mean())

# Q12
print("Q12:\n", df.sort_values(['A','B'], ascending=False))

# Q13
df3 = pd.DataFrame({'A':[1,2],'C':[7,8]})
print("Q13:\n", pd.merge(df, df3, on='A'))

# Q14
df4 = pd.DataFrame({'A':[1,None,3]})
print("Q14:\n", df4.fillna(0), "\n\n", df4.dropna())

# Q15
print("Q15:\n", df2['grp'].value_counts())

# Q16
print("Q16:", dict(zip(df['A'], df['B'])))

# Q17
print("Q17:")
for i,row in df.iterrows():
    print(i, row['A'], row['B'])

# Q18
df['index'] = [i for i,_ in enumerate(df['A'])]
print("Q18:\n", df)

# Q19
df['A'] = list(map(lambda x: x+1, df['A']))
print("Q19:\n", df)

# Q20
filtered = list(filter(lambda x: x>2, df['A']))
print("Q20:", filtered)
````
