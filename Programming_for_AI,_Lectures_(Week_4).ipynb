### **NumPy** Library: The Core of Scientific Computing


- NumPy (Numerical Python) is crucial for efficient numerical operations in
machine learning.
- It provides powerful array objects and tools for working
with them.

- Arrays are faster and more memory-efficient than Python lists
for numerical computations.


         Let's create NumPy arrays, also known as vectors (1D arrays) and matrices (2D arrays), and perform some basic operations.

```python
import numpy as np

# --- Creating NumPy Arrays ---

# **Creating a 1D Array (Vector)**
# A 1D array is simply a sequence of numbers, often representing features or data points.


vector_a = np.array([5, 10, 15, 20])
vector_b = np.array([10, 20, 30, 40])

print(f"Vector A: {vector_a}")
print(f"Vector B: {vector_b}")
```

```python
# Example: Creating a vector of zeros
zeros_vector = np.zeros(5)
print(f"Vector of zeros: {zeros_vector}")

# Example: Creating a vector with a range of values
range_vector = np.arange(0, 10, 2) # Start, Stop (exclusive), Step
print(f"Vector with range: {range_vector}")
```

### **Creating a 2D Array (Matrix)**
- A 2D array, or matrix, is a grid of numbers with rows and columns.
- They are essential for representing datasets, where rows are samples and columns are features.

```python
matrix_A = np.array([  [1, 2] , [3, 4] ])
print("\nMatrix A:")
print(matrix_A)

# Example: Creating a 3x3 identity matrix
identity_matrix = np.eye(3)
print("\nIdentity Matrix (3x3):")
print(identity_matrix)

# Example: Creating a matrix of ones
ones_matrix = np.ones((2, 3)) # (rows, columns)
print("\nMatrix of ones (2x3):")
print(ones_matrix)
```

```python
# Basic Array Operations ---

# **Element-wise Addition**
# This operation adds corresponding elements of two arrays.
vector_sum = vector_a + vector_b
print(f"\nSum of vectors (element-wise): {vector_sum}")

# Example: Element-wise multiplication
vector_product = vector_a * vector_b
print(f"Product of vectors (element-wise): {vector_product}")
```

```python
# **Dot Product (Matrix Multiplication)**
# The dot product is a fundamental operation in linear algebra, widely used in ML.
# For two vectors `[a1, a2, a3]` and `[b1, b2, b3]`, the dot product is `a1*b1 + a2*b2 + a3*b3`.
dot_product = np.dot(vector_a, vector_b)
print(f"Dot product of vectors: {dot_product}")

# Example: Matrix multiplication
matrix_B = np.array([3,4])
matrix_product = np.dot(matrix_A, matrix_B)
print("\nMatrix A * Matrix B (dot product):")
print(matrix_product)
```

### **Pandas**: Data Analysis Made Easy

- Pandas is a powerhouse for data manipulation and analysis, especially when dealing with structured data.
- Its primary data structure, the **DataFrame**, is incredibly versatile, resembling spreadsheets or SQL tables.

```python
# Creating DataFrames ---

import pandas as pd

# **Creating a Simple DataFrame from a Dictionary**
# DataFrames can be created from various sources, including Python dictionaries,
# lists of dictionaries, or by reading files like CSVs.
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 40],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
print("\nPandas DataFrame from dictionary:")
display(df)

# Example: Creating a DataFrame from a list of lists
data_list = [['David', 40, 'Houston'],
             ['Eve', 28, 'Miami']]
df_list = pd.DataFrame(data_list, columns=['Name', 'Age', 'City'])
print("\nPandas DataFrame from list of lists:")
display(df_list)
```

```python
# Accessing Data in DataFrames ---

# **Accessing a Specific Column**
# You can easily select a column by its name, which returns a Pandas Series (a 1D labeled array).
print("\nAges column:")
display(df['Age'])

# Example: Accessing multiple columns
print("\nName and City columns:")
display(df[['Name', 'City']])

# Example: Accessing rows by index (loc and iloc)
print("\nFirst row using .iloc (integer-location based indexing):")
display(df.iloc)

print("\nRows for Alice and Charlie using .loc (label-based indexing):")
display(df.loc[1])

# Example: Filtering rows based on a condition
print("\nPeople older than 28:")
display(df[df['Age'] > 28])
```

```python
# Adding/Modifying Columns ---

# Example: Adding a new column
df['Occupation'] = ['Engineer', 'Artist', 'Doctor']
print("\nDataFrame after adding 'Occupation' column:")
print(df)

# Example: Modifying an existing column
df['Age'] = df['Age'] + 1 # Increment all ages by 1
print("\nDataFrame after incrementing 'Age':")
print(df)
```

```python
# Let's create 1D arrays (vectors) and perform element-wise addition and a dot product.
vector_a = np.array([-9, 8, 45]) #
vector_b = np.array([3, -7, 2]) #

print(f"Vector A: {vector_a}") #
print(f"Vector B: {vector_b}") #

# Element-wise addition
vector_sum = vector_a + vector_b #
print(f"Sum of vectors: {vector_sum}") #

# Dot product
dot_product = np.dot(vector_a, vector_b) #
print(f"Dot product of vectors: {dot_product}") #
```

###  Matrices
- A matrix is a 2D array of numbers. They are crucial for representing transformations and complex datasets.



```python
# Create a 2D array (matrix)
matrix_A = np.array([    [4, -9],[5, 7]   ]) #
matrix_B = np.array([    [0, 5],[6, -8]   ]) #

print("\nMatrix A:") #
print(matrix_A) #
print("\nMatrix B:") #
print(matrix_B) #

```

```python
# Matrix addition is performed element-wise.

matrix_C = matrix_A + matrix_B #
print(f"\nMatrix A + B:\n{matrix_C}") #

# **Example 1.2.3: Matrix Multiplication (Dot Product)**
#
# Matrix multiplication is not element-wise. It follows specific rules of row-column dot products.

matrix_D = np.dot(matrix_A, matrix_B) #
print(f"\nMatrix A * B:\n{matrix_D}") #
```

```python
# Determinant and Inverse of a Matrix
## The **determinant** is a scalar value that provides insights into a square matrix, such as whether it's invertible. The **inverse** of a matrix, if it exists, is another matrix that when multiplied by the original, yields the identity matrix.
## **Note:** The inverse only exists for square matrices with a non-zero determinant.

determinant_A = np.linalg.det(matrix_A) #
inverse_A = np.linalg.inv(matrix_A) #
print(f"\nDeterminant of A: {determinant_A:.2f}") #
print(f"\nInverse of A:\n{inverse_A}") #

# Let's verify the inverse: A * A_inverse should be close to the identity matrix
identity_check = np.dot(matrix_A, inverse_A)
print(f"\nVerification (A * Inverse_A):\n{identity_check}")

# Example of a singular matrix (determinant is 0, so no inverse)
singular_matrix = np.array([[1,2], [2,4]])
determinant_singular = np.linalg.det(singular_matrix)
print(f"\nDeterminant of a singular matrix:\n{singular_matrix}")
print(f"Determinant: {determinant_singular:.2f}")
try:
    np.linalg.inv(singular_matrix)
except np.linalg.LinAlgError as e:
    print(f"Error trying to compute inverse of singular matrix: {e}")

```

```python
# Transpose of a Matrix
## The transpose of a matrix is obtained by flipping the matrix over its diagonal, i.e., interchanging the rows and columns.

matrix_F = np.array([1,2,3,4,5,6]).reshape([2,3])
print(f"\nOriginal Matrix F:\n{matrix_F}")

matrix_F_transpose = matrix_F.T
print(f"Transpose of Matrix F:\n{matrix_F_transpose}")

```

### **Calculus**: The Study of Change
- Calculus, particularly **derivatives** (or gradients), is vital in AI and for training machine learning models.
- Algorithms like gradient descent use derivatives to find the optimal parameters that minimize a model's error.
- We can use NumPy to approximate gradients.


```python
### Approximating Gradients (Derivatives)
## The `np.gradient()` function numerically approximates the derivative of a function.

# **Example 2.1.1: Gradient of a Quadratic Function**
# # We'll use the example provided in the source: `f(x) = x^2 + 2x + 1`. The true derivative is `f'(x) = 2x + 2`.

# Define a simple quadratic function
def f(x): #
    return x**2 + 2*x + 1 #

# Generate x values
x = np.linspace(-10, 10, 100) #
y = f(x) #

# Calculate the gradient (derivative)
gradient_f = np.gradient(y, x) #

# Plot the function and its gradient
plt.figure(figsize=(10, 6)) #
plt.plot(x, y, label='f(x) = x^2 + 2x + 1') #
plt.plot(x, gradient_f, label="f'(x) - Gradient (Numerical)", linestyle='--') #
plt.title('Function and its Gradient') #
plt.xlabel('x') #
plt.ylabel('y') #
plt.legend() #
plt.grid(True) #
plt.show() #
```

```python
# Gradient of a Cubic Function
## Let's explore another function: `g(x) = x^3 - 6x^2 + 9x + 1`. The true derivative is `g'(x) = 3x^2 - 12x + 9`.


# Define a cubic function
def g(x):
    return x**3 - 6*x**2 + 9*x + 1

# Generate x values
x_g = np.linspace(-2, 5, 100)
y_g = g(x_g)

# Calculate the gradient
gradient_g = np.gradient(y_g, x_g)

# Plot the function and its gradient
plt.figure(figsize=(10, 6))
plt.plot(x_g, y_g, label='g(x) = x^3 - 6x^2 + 9x + 1')
plt.plot(x_g, gradient_g, label="g'(x) - Gradient (Numerical)", linestyle='--')
plt.title('Cubic Function and its Gradient')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Gradient of a Trigonometric Function
## Consider `h(x) = sin(x)`. The true derivative is `h'(x) = cos(x)`.

# Define a trigonometric function
def h(x):
    return np.sin(x)

# Generate x values
x_h = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y_h = h(x_h)

# Calculate the gradient
gradient_h = np.gradient(y_h, x_h)

# Plot the function and its gradient
plt.figure(figsize=(10, 6))
plt.plot(x_h, y_h, label='h(x) = sin(x)')
plt.plot(x_h, gradient_h, label="h'(x) - Gradient (Numerical)", linestyle='--')
plt.title('Sine Function and its Gradient')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

### **Probability**: Quantifying Uncertainty
- Probability theory is essential for understanding uncertainty, making predictions, and evaluating models in AI and machine learning.
- It helps us quantify the likelihood of events.



```python
### Basic Probability
## Let's start with some fundamental concepts like the probability of simple events.

# Probability of Rolling an Even Number on a Die
## A standard six-sided die has outcomes {1, 2, 3, 4, 5, 6}. Even numbers are {2, 4, 6}.

total_outcomes_die = 6
favorable_outcomes_even = 3 # (2, 4, 6)
probability_even = favorable_outcomes_even / total_outcomes_die
print(f"Probability of rolling an even number: {probability_even:.2f}")

# **Example 3.1.2: Probability of Drawing a Red Card from a Deck**
## A standard deck of 52 cards has 26 red cards (Hearts and Diamonds).

total_cards = 52
red_cards = 26
probability_red_card = red_cards / total_cards
print(f"Probability of drawing a red card: {probability_red_card:.2f}")
```

```python
### Bayes' Theorem
## **Bayes' Theorem** is crucial for updating probabilities based on new evidence. It allows us to calculate the **posterior probability** of an event (P(A|B)) given **prior probabilities** and **likelihoods**.
## The formula is: **P(A|B) = (P(B|A) * P(A)) / P(B)**.

# **Example 3.2.1: Disease Testing Scenario**
## This classic example demonstrates how Bayes' Theorem helps interpret medical test results, highlighting the importance of the **base rate (prevalence)** of a disease.
#
# **Scenario:** A test for a rare disease.
# *   `P(A)` = Probability of actually having the disease (prevalence) = **1%**.
# *   `P(Not A)` = Probability of not having the disease = **99%**.
# *   `P(B|A)` = Probability of testing positive **given** you have the disease (Sensitivity) = **99%**.
# *   `P(B|Not A)` = Probability of testing positive **given** you do NOT have the disease (False Positive Rate) = **5%**.
#
# We want to find `P(A|B)`: The probability of having the disease given a positive test result.


# P(A) = P(has disease) = 1%
p_disease = 0.01 #
# P(Not A) = P(does not have disease) = 99%
p_no_disease = 0.99 #

# P(B|A) = P(test is positive | has disease) = 99% (Sensitivity)
p_pos_given_disease = 0.99 #
# P(B|Not A) = P(test is positive | does not have disease) = 5% (False Positive Rate)
p_pos_given_no_disease = 0.05 #

# First, calculate the total probability of testing positive: P(B)
# P(B) = P(B|A)*P(A) + P(B|Not A)*P(Not A)
p_positive = (p_pos_given_disease * p_disease) + \
             (p_pos_given_no_disease * p_no_disease) #
print(f"Total probability of testing positive (P(B)): {p_positive:.4f}") #

# Now, use Bayes' Theorem to find P(A|B): Probability of having the disease given a positive test.
p_disease_given_positive = (p_pos_given_disease * p_disease) / p_positive #
print(f"\nProbability of having the disease given a positive test (P(A|B)): "
      f"{p_disease_given_positive*100:.2f}%") #

print("\nThis surprising result highlights the importance of considering the base rate (prevalence of the disease)." #
      " Even with a high sensitivity test, if the disease is rare, a positive result doesn't guarantee you have it due to false positives.")
```

```python
# Spam Email Classification
## Let's consider a simpler Bayes' theorem example: Classifying an email as spam.
#
# *   `P(Spam)` = Probability that an email is spam = 20%
# *   `P(Not Spam)` = Probability that an email is not spam = 80%
# *   `P("Make money fast" | Spam)` = Probability of the word "Make money fast" appearing in a spam email = 10%
# *   `P("Make money fast" | Not Spam)` = Probability of "Make money fast" appearing in a non-spam email = 0.1%
#
# We want to find `P(Spam | "Make money fast")`: The probability that an email is spam given it contains the word "Make money fast".

p_spam = 0.20
p_not_spam = 0.80

p_Make_money_fast_given_spam = 0.10
p_Make_money_fast_given_not_spam = 0.001

# Calculate P("Make_money_fast") = P("Make_money_fast" | Spam)*P(Spam) + P("Make_money_fast" | Not Spam)*P(Not Spam)
p_Make_money_fast = (p_Make_money_fast_given_spam * p_spam) + \
           (p_Make_money_fast_given_not_spam * p_not_spam)
print(f"Total probability of an email containing 'Make_money_fast' (P('Make_money_fast')): {p_Make_money_fast:.4f}")

# Calculate P(Spam | "Make_money_fast")
p_spam_given_Make_money_fast = (p_Make_money_fast_given_spam * p_spam) / p_Make_money_fast
print(f"Probability of an email being spam given it contains 'Make_money_fast': "
      f"{p_spam_given_Make_money_fast*100:.2f}%")

print("\nAs seen, even if the word 'Make_money_fast' is rare in general emails, its presence significantly increases the probability that an email is spam due to its much higher likelihood in spam messages.")
```

     - create random data
     - store in DataFrame
     - filter values
     - compute stats
     - visualize results





```python
# =========================
# SOLUTIONS
# =========================

print("\n\nSOLUTIONS:\n")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Q1
arr = np.array([1,2,3,4,5])
print("Q1:", arr.mean(), arr.max(), arr.min())

# Q2
arr = np.arange(1,21).reshape(4,5)
print("Q2:\n", arr)

# Q3
print("Q3:")
print(arr[0])
print(arr[:,-1])
print(arr[1:3,1:4])

# Q4
arr = np.arange(10)
arr[arr%2==0] = 0
print("Q4:", arr)

# Q5
def square(arr):
    return arr**2
print("Q5:", square(np.array([1,2,3])))

# Q6
a = np.array([1,2,3])
b = np.array([4,5,6])
print("Q6:", a+b, a*b, np.dot(a,b))

# Q7
arr = np.array([10,20,30,40])
print("Q7:", arr[arr > arr.mean()])

# Q8
df = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
print("Q8:\n", df.describe())

# Q9
print("Q9:\n", df[df['A']>1])

# Q10
df['C'] = df['A'].apply(lambda x: x*2 if x>1 else x)
print("Q10:\n", df)

# Q11
df2 = pd.DataFrame({'grp':['A','A','B'],'val':[10,20,30]})
print("Q11:\n", df2.groupby('grp')['val'].mean())

# Q12
print("Q12:\n", df.sort_values(['A','B'], ascending=False))

# Q13
df3 = pd.DataFrame({'A':[1,2],'C':[7,8]})
print("Q13:\n", pd.merge(df, df3, on='A'))

# Q14
df4 = pd.DataFrame({'A':[1,None,3]})
print("Q14:\n", df4.fillna(0), "\n\n", df4.dropna())

# Q15
print("Q15:\n", df2['grp'].value_counts())

# Q16
print("Q16:", dict(zip(df['A'], df['B'])))

# Q17
print("Q17:")
for i,row in df.iterrows():
    print(i, row['A'], row['B'])

# Q18
df['index'] = [i for i,_ in enumerate(df['A'])]
print("Q18:\n", df)

# Q19
df['A'] = list(map(lambda x: x+1, df['A']))
print("Q19:\n", df)

# Q20
filtered = list(filter(lambda x: x>2, df['A']))
print("Q20:", filtered)

# Q21
arr = np.arange(10)
plt.figure()
plt.plot(arr)
plt.title("Q21 Line Plot")
plt.show()

# Q22
plt.figure()
plt.plot(arr, label='line1')
plt.plot(arr*2, label='line2')
plt.legend()
plt.show()

# Q23
df.plot(kind='bar')
plt.title("Q23 Bar Chart")
plt.show()

# Q24
plt.hist(np.random.randn(100))
plt.title("Q24 Histogram")
plt.show()

# Q25
x = np.arange(10)
y = x**2
plt.scatter(x,y)
plt.title("Q25 Scatter")
plt.show()

# Q26
plt.figure()
for i in range(3):
    plt.plot(np.arange(10), np.arange(10)*(i+1))
plt.title("Q26 Multiple Lines")
plt.show()

# Q27
df.plot()
plt.title("Q27 DataFrame Plot")
plt.show()

# Q28
plt.figure()
plt.plot(x,y)
plt.title("Custom Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()

# Q29
arr = np.random.randint(1,10,(5,3))
df = pd.DataFrame(arr, columns=['A','B','C'])
df.plot()
plt.title("Q29 NumPy → Pandas → Plot")
plt.show()

# Q30
data = np.random.randint(1,100,20)
df = pd.DataFrame({'vals':data})
filtered = df[df['vals']>50]
print("Q30 Stats:", filtered.describe())
filtered.plot(kind='hist')
plt.show()
```
