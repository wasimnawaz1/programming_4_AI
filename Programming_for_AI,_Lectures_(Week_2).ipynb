### `ord()` and `chr()` functions
In Python, characters are stored as Unicode. For standard English letters and symbols, these align exactly with the ASCII table.
- Unicode is a superset of ASCII, maintaining backward compatibility.
- `ord(char)`: Takes a single character and returns its integer representation.
- `chr(code)`: Takes an integer and returns its character representation.

````python
text = "Hello!"

for char in text:
    ascii_val = ord(char)
    print(f"Character: {char} | ASCII: {ascii_val}")


ASCII_numbers = [80, 121, 116, 104, 111, 110]

# Get a list of ASCII integers
ascii_char_list = [chr(number) for number in ASCII_numbers]

print(f"Original ASCII: {ASCII_numbers}")
print(f"Characters: {ascii_char_list}")

# Using an empty string as the separator
string_from_char_list = "".join(ascii_char_list)

print(f"String from characters: {string_from_char_list}")
````

## String Operations — Accessing Elements & Slicing

Python allows **positive** and **negative** indices to access elements of a string.

- Indexing starts from `0`
- Negative index `-1` refers to the last character

**String Slicing:** `x[p1:p2]` returns characters from position `p1` up to (not including) `p2`.

**Stride (Step):** `x[p1:p2:step]` retrieves every `step`-th character.

````python
x = "Hello world"
print(x[0])   # First character
print(x[-1])  # Last character
````

````python
print(x[0:5])           # Characters from index 0 to 4
print(x[6:len(x)])      # Characters from index 6 to end
````

````python
x = 'Hello world'
x[:6] + x[6:] == x
````

````python
# Stride (Step) in String Slice
x = 'Hello world'
x[1:8:2]  # every second character from index 1 to 7
````

````python
x = 'Hello world'
x[10:0:-2]  # step backward
````

````python
x = 'Hello world'
x[10::-2]  # step backward, including index 0
````

````python
# Replicate a string using * operator
str = '123' * 3
str
````

````python
str = '123' * 3
str[: : -3]  # every third element stepping backward
````

## Dealing with Conditional Statements and Iterations

To control the flow of a program, Python provides conditional statements:

1. `if`
2. `elif`
3. `else`

**Syntax of `if` statement:**
```python
if(condition):
    Statement1
```
- Statements following `if` are **indented**.
- Executed only when the condition is `True`.

````python
obtained_marks = 80
if(obtained_marks > 100):
    print("Invalid marks.")
````

````python
print('Input marks of a student')
obtained_marks = int(input())
if(obtained_marks > 100):
    print("Invalid marks.")
````

## `else` Statement

The `else` statement always comes with an associated `if`:

```python
if(condition):
    # statements when condition is True
else:
    # statements when condition is False
```

````python
print('Input marks of a student')
obtained_marks = int(input())
if(obtained_marks >= 80):
    print("Excellent")
else:
    print("Not excellent")
````

## Nested Decisions — `if-elif-else` Statement

For complex decisions with multiple conditions:

```python
if(condition1):
    # ...
elif(condition2):
    # ...
else:
    # ...
```

````python
print('Input marks of a student')
obtained_marks = int(input())
if(obtained_marks > 100 or obtained_marks < 0):
    print("Invalid marks.")
elif(obtained_marks >= 80):
    print("Excellent")
else:
    print("Not excellent")
````

## For Loop

A `for` loop is used when we need to perform a task a **specific number of times**.

**Syntax:**
```python
for loop_variable in input_variable:
    Statement(s) to be executed in the for loop
```

- The number of iterations depends on the length of `input_variable`.
- `range(n)` generates numbers from `0` to `n-1`.
- `break` — exits the loop.
- `continue` — skips the rest of the current iteration.

````python
fruits = ["pineapple", "orange", "banana"]
for k in fruits:
    print(k)
````

````python
# range() in a for loop
for j in range(10):
    print(j)
````

````python
# range with step size
for x in range(2, 10, 3):
    print(x)
````

````python
# Using break and continue
fruits = ["pineapple", "orange", "banana", "melon", "dates"]
for k in fruits:
    if k == "banana":
        continue
    print(k)
    if k == "melon":
        break
````

````python
# for loop with else
numbers = [0, 1, 2, 3, 4, 5]
for i in numbers:
    print(i)
else:
    print("No items left.")
````

````python
### Checking whether a target item found or not
numbers = [10, 25, 30, 45, 50]
target = 40

for item in numbers:
    if item == target:
        print("Found!")
        break
else:
    print("Not found")
````

## While Loop

The `while` loop iteratively executes code **until a condition is met**.

**Syntax:**
```python
while (condition):
    Statement(s) to be executed in the while loop
```

Similar to `for`, it also supports `break`, `continue`, and `else`.

````python
# Program to add natural numbers provided by the user.
n = int(input("Enter an integer = "))

# initialize sum to 0 and a variable k as a counter
sum = 0
k = 1
while k <= n:
    sum = sum + k
    k = k + 1  # update counter
print("The sum of first", n, "natural number is=", sum)
````

## Nested Loops

Both `for` and `while` loops can be **nested** inside each other — useful for 2-dimensional operations.

````python
properties = ["red", "round", "tasty"]
fruits = ["apple", "orange", "banana"]
for j in properties:
    for k in fruits:
        print(j, k)
````
