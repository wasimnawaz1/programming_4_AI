## Traditional AI programming languages
- **Lisp** (List Processing): Developed in 1958 by John McCarthy,
    - Lisp is considered the first, and historically an important, AI programming language.
    - It was designed for symbolic processing, which is well-suited for AI tasks that involve manipulating rules and structures rather than numbers.
    - Its ability to treat code as data allowed flexible AI systems that could adapt during runtime.
- **Prolog** (Programming in Logic): Emerged in the 1970s.
    - A declarative language that defines "what" the problem is through facts and rules rather than the step-by-step "how" of traditional imperative programming.
    - It excels in symbolic reasoning, natural language processing, and knowledge representation, making it a cornerstone for early expert systems.
- **Smalltalk**: Developed at Xerox PARC in the 1980s.
     - Smalltalk was foundational in introducing Object-Oriented Programming (OOP) concepts, which are now widely used in designing modular, complex AI systems.
 - **CLIPS** (C Language Integrated Production System), developed by NASA's Johnson Space Center from 1985 to 1996.
     - A specialized tool created to build rule-based expert systems, which aim to solve complex problems by reasoning using `if–then` rules rather than through conventional procedural programming code.

## Python's Approach (Hybrid Model)
Python is often referred to as an **interpreted language**, but its implementation, CPython, uses a hybrid approach.
- When a Python program runs, the source code is first compiled into an intermediate representation called **bytecode**.
- This bytecode is then executed by a **Python Virtual Machine** (PVM), which acts as the interpreter.
- The initial compilation step happens automatically and is largely hidden from the user, which is why Python feels like an interpreted language in practice.

Other languages like C and C++ are compiled, while languages like Ruby and JavaScript are interpreted.

### Raw vs. Formatted Strings

Strings can be handled to:
- treat special characters (like backslashes) and
- inject variables into text

#### 1. Raw Strings (`r"..."`)

Python uses the backslash (`\`) as an **escape character** (e.g., `\n` for a new line, `\t` for a tab).

* **The Problem:** When dealing with Windows file paths or Regular Expressions (Regex), backslashes are everywhere. Having to type `\\` for every single slash is tedious and messy.
* **The Solution:** A **Raw String** tells Python: "Ignore all escape sequences. Treat backslashes as literal characters."

#### 2. Formatted Strings (`f"..."`)

Introduced in Python 3.6, **f-strings** (formatted string literals) are the modern, fastest, and most readable way to perform string interpolation.

- By prefixing a string with `f`, we can embed Python expressions directly inside curly braces `{}`.



````python
# Normal String: \n is treated as a newline
normal_path = "C:\new_folder\tasks"
print("Normal String Output:")
print(normal_path)

print("-" * 20)

# Raw String: \n is treated as literal text
raw_path = r"C:\new_folder\tasks"
print("Raw String Output:")
print(raw_path)
````

````python
### Dynamic Text with F-Strings, which allow us to perform calculations, call methods, and format numbers inside the string.

name = "Alice"
age = 25
pi = 3.14159265

# 1. Simple variable injection
print(f"Hello, my name is {name} and I am {age} years old.")

# 2. Expressions and methods inside braces
print(f"Next year, I will be {age + 1}. My name in caps is {name.upper()}.")

# 3. Number formatting (limiting decimals)
print(f"The value of Pi to two decimal places is {pi:.2f}.")
````

````python
### Combining Raw and Formatted Strings to handle Regex patterns that include variables (`fr"..."`).

folder_name = "user_data"

# Combining both: raw to handle slashes, f to handle the variable
complex_path = fr"C:\{folder_name}\reports\daily_log.txt"

print(complex_path)
````

## Bitwise Operators

Bitwise operators operate on individual bits of the operands.

- Every number is represented as a series of 0s and 1s in the computer.
- For example, decimal 5 = `0000 0101` in binary (using 8 bits).
- Negative numbers use **2's complement** representation.

| Operator | Name | Example (x=2, y=5) |
|----------|------|---------------------|
| `&` | Bitwise AND | `x & y` → `0` |
| `\|` | Bitwise OR | `x \| y` → `7` |
| `~` | Bitwise NOT | `~x` → `-3` |
| `^` | Bitwise XOR | `x ^ y` → `7` |
| `>>` | Right Shift | `y >> 2` → `1` |
| `<<` | Left Shift | `x << 2` → `8` |

````python
# To take negative of number 3 in Python:
~3 + 1
````

````python
x = 2  # 0000 0010 in binary
y = 5  # 0000 0101 in binary

print("x & y =", x & y)   # Bitwise AND
print("x | y =", x | y)   # Bitwise OR
print("~x =", ~x)          # Bitwise NOT
print("x ^ y =", x ^ y)   # Bitwise XOR
print("y >> 2 =", y >> 2)  # Right Shift
print("x << 2 =", x << 2)  # Left Shift
````

## Logical Operators

Logical operators work on `True` and `False` operands.

| Operator | Description | Example |
|----------|-------------|---------|
| `and` | Returns True only when both operands are True | `True and False` → `False` |
| `or` | Returns True if either operand is True | `True or False` → `True` |
| `not` | Complements the operand | `not True` → `False` |

````python
x = True
y = False

print(x and y)
print(x or y)
print(not x)
````

## Comparison Operators

These operators compare two values and return `True` or `False`. They are extensively used to make decisions in programming.

| Operator | Description | Example (x=8, y=3) |
|----------|-------------|---------------------|
| `==` | Equal | `x == y` → `False` |
| `!=` | Not equal | `x != y` → `True` |
| `>` | Greater than | `x > y` → `True` |
| `<` | Less than | `x < y` → `False` |
| `>=` | Greater than or equal | `x >= y` → `True` |
| `<=` | Less than or equal | `x <= y` → `False` |

````python
x = 8
y = 3

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
````

## Membership Operators

````python
x = 'Hello world'  # x is a string of characters
print('H' in x)           # returns True if 'H' is in x
print('Hello' not in x)   # returns True if 'Hello' is not present in x
````
